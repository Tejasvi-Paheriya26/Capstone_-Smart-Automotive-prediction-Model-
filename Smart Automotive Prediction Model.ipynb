## Import Libraries

```python
import pandas as pd                  #For Handle data
import numpy as np                   # for numerical operations  
import matplotlib.pyplot as plt      # use for basic plotting
import seaborn as sns                # Advanced data visualization


from sklearn.model_selection import train_test_split    # data split into train
from sklearn.linear_model import LogisticRegression     # ML(Linear Regression ) model
from sklearn.metrics import accuracy_score              # check model accuracy
```

## Load Dataset

                 

```python
df = pd.read_csv("Smart Automotive Prediction.csv")     # load dataset    
df.head()                                               # print the first 5 rows
```

## Pairplot(Relation Between features)


```python
sns.pairplot(df)                # use target column for color
plt.show()
```

## Heatmap (correlation matrix)


```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

## Define features (X) and Target (y)

```python

X = df[['Mileage_km', 'Engine_Temp', 'Vibration', 'Oil_Level']]
y = df['Maintenance_Needed']      # 0 = no, 1  = yes
```

## Train Machine Learning Model

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, random_state = 42
)
```

## Prediction & Evalution


```python
model = LogisticRegression(max_iter = 1000)  # 100 default time iteration
model.fit (X_train, y_train)
```

## New vehicle Prediction


```python
New_vehicle = [[600000, 95, 0.8, 58]]

prediction = model.predict(New_vehicle)

if prediction[0] == 1:
    print(" Maintenance Needed")
else:
    print(" No maintenance Needed")
```
